1. Förklara övergripande vad NoSQL är.


Svar:

NoSQL är en typ av databas där man kan lagra data som inte är helt relational, det kan stå för "Non SQL" eller "Not Only SQL" eftersom idag finns det
många databastjänster som kan hantera både strukturerad relationell data och semi-strukturerad data i samma databas, samt databastjänster där man kan använda SQL-liknande syntax för frågeställning fast datan inte är helt strukturerad. NoSQL kan i stort sätt delas upp i "nyckel/värde" databaser, "dokument" databaser, "bredkolumn" databaser och "graph" databaser. 

2. Vilken typ av data kan en NoSQL-databas hantera som inte en
relationell databas kan?


Svar: 

NoSQL kan hantera semi och ostrukterad data. Semi-strukturerad data är typ av data där alla eniteter och instancer inte nödvändigt har samma egenskaper och förhållande, det gör det svårt att få den typen data i en relationell databas. I semi-strukturerad databas kan man använda nycklar för att peka på olika instancer men man kan inte använda foreign keys för att peka till andra tabeller. Ostrukturerad data är typen av data som inte har nån struktur alls, det kan vara bilder, ljudinspelningar, video, binära filer osv. 

3. Beskriv kortfattat vad en nyckel/värde-databas är.


Svar: 
Nyckel/värde databas är en enkel databas där man kan lagra nyckel/värde parer. I en nyckel/värde databas har man inte möjlighet att söka i den igenom nåt annat en nyckeln (oftast kanske ett id på en instans). Den typen databas är ganska snabb och enkel, den är lätthanterad och tar inte mycket minne eller kraft att driva. Värden i en nyckel/värde databas kan vara olika och det går att ändra värden om det behövs (fast man inte har tillgång til värdena om man inte har nyckeln).

4. Beskriv kortfattat vad en dokumentdatabas är.


Svar:

Dokument databas är i grunden en typ av nyckel/värde databas fast med utökat stöd för queries och olika datatyper. En dokument databas består av grupperingar av "dokument", i collections, där vart dokument innehåller olika nyckel/värde par. Dokument databas kräver inte att man har en fördefinerad struktur, det möjligör att dokument kan ha olika innehåll även om de sitter i samma collection (ett collection består kanske av kundinformation men olika kunder kan sen ha olika mängder information eller olika typer data i sitt dokument). 

5. Beskriv kortfattat vad en bredkolumndatabas är.


Svar:

Bredkolumndatabas är en databas av tabeller och kan på ytan se ut som en relationell databas. Stor skillnad är då att i en bredkolumndatabas behöver inte alla rader att ha samma kolumner och när kan skapar en kolumn då skapar man den bara för de raderna som man vill. Rader i en bredkolumndatabas som innehåller samma kolumner kan grupperas in kolumn-familjer som gör det snabbt att söka i databasen eftersom man söker bara i de raderna som innehåller de kolumnerna man är ute efter. 



6. Beskriv kortfattat vad en grafdatabas är.

Svar:

I en grafdatabas lagrar man data i noder och kanter, där noder innehåller enheter och kanter innehåller relationer mellan enheterna. Kanter kan vara riktade, där de pekar från en nod til en annan, eller oriktade där de visar gemensamma relationer. Grafdatabaser är väldigt bra och användbara om man vill kunna visa och visualisera relationer mellan olika enheter. Fast grafdatabas är inte relationell i sin struktur, man behöver inte ha en definerad struktur och man kan lägga til noder och kanter utan att behöva ta hänsyn till vad som redan finns i databasen.

I mappen `kunskapskontroll_1/data/northwind/` finns filerna `products.csv` och `suppliers.json`.

Slå ihop `products.csv` och `suppliers.json` på ett passande sätt och spara dem som en dokumentdatabas i MongoDB.

#### Godkänt
Ta reda på:
* vilka produkter som behöver beställas (`ReorderLevel` $\gt$ `UnitsInStock` $ + $ `UnitsOnOrder`)
* vem man ska kontakta för att beställa fler, och
* hur man kontaktar personen


```python
df = pd.read_csv(r"data\northwind\products.csv")
df1 = pd.read_json("data/northwind/suppliers.json")
```

```python
df3 = pd.merge(df, df1, on='SupplierID')
```

```python
data = json.loads(
    df3.reset_index()
    .drop("index", axis=1)
    .rename({"ProductID": "_id"}, axis=1)
    .to_json(orient="records")
    )
data
```

```python
from pymongo.mongo_client import MongoClient
from pymongo.server_api import ServerApi

PWD = open("mongodb.pwd", "r").read().strip()

uri = f"mongodb+srv://vidar1812:{PWD}@cluster0.zaamq.mongodb.net/"

# Create a new client and connect to the server
client = MongoClient(uri, server_api=ServerApi('1'))

# Send a ping to confirm a successful connection
try:
    client.admin.command('ping')
    print("Pinged your deployment. You successfully connected to MongoDB!")
except Exception as e:
    print(e)
```

```python
database = client['Database']

collection = database['Products']
```

```python
collection.insert_many(data)
```

```python
for d in collection.find():
    print(d)
```

```python
results = collection.aggregate([
    {
        '$addFields': {
            'StockOrders': {
                '$add': [
                    '$UnitsInStock', '$UnitsOnOrder'
                ]
            }
        }
    }, {
        '$match': {
            '$expr': {
                '$gt': [
                    '$ReorderLevel', '$StockOrders'
                ]
            }
        }
    }
])
```

```python
for res in results:
    print(res)
```
